```python
import pandas as pd

df = pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

Here the missing values are replaced with '0' using isna()

```python
df.drop_duplicates(inplace=True)
```

```python
text=["restaurant name", "restaurant type","cuisines type","area","local address"]
for x in text:
  df[x]=df[x].str.strip().str.title().str.lower()
```

Created a list so that weecan easily clean all the rrequired columns

```python
df=df.convert_dtypes()
```

```python
df.rename(columns={
    'Unnamed: 0.1'       : 'index_1',
    'Unnamed: 0'         : 'index_2',
    'restaurant name'    : 'restaurant_name',
    'restaurant type'    : 'restaurant_type',
    'rate (out of 5)'    : 'rating',
    'num of ratings'     : 'num_ratings',
    'avg cost (two people)' : 'avg_cost_two',
    'table booking'      : 'table_booking',
    'cuisines type'      : 'cuisines_type',
    'local address'      : 'local_address',
}, inplace=True)
```

```python
df['hybrid'] = (df['online_order'] == 'Yes') & (df['table_booking'] == 'Yes')
```

Created a new column to see whether restuarants offer both online &offline mode of transportation or not

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
